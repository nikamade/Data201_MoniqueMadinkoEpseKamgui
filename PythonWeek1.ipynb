Part A -- Translation (Conceptual)
For the R code below:

Predict the output\
Translate it into Python\
Compute the result in Python
x <- c(1, 3, 5, 7)
y <- ifelse(x > 4, x^2, x + 1)
mean(y)

Predticted OutPut Result: 20

```python
import numpy as np
x = np.array([1,3, 5, 7])
y = np.where(x > 4, x**2, x + 1)
np.mean(y)

```

Part B -- Lists vs NumPy Arrays
Create a Python list of numbers 1--5\
Attempt to multiply the list by 2\
Convert the list to a NumPy array and repeat the multiplication
Answer the following questions:

What happened in each case?\
- in the list is multiplied by 2 so the list is repeated.
- with numpy, the every element in the list is multiplied by 2

Why does NumPy behave differently?


```python
my_list = [1,2,3,4,5]
multiplied_list = my_list * 2
print(multiplied_list)
```

```python
import numpy as np
my_arraylist = np.array([1,2,3,4,5])
multiplied_arraylist = my_arraylist * 2
print("multiplied_arraylist:", multiplied_arraylist)
```
